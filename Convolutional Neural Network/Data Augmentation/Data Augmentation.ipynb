# <center>Data Augmentation</center>

```python
from keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.preprocessing import image
from tensorflow.keras.preprocessing.image import img_to_array,img_to_array, load_img
```

# 1. Single Image

```python
'''
Loading Image
'''
img=image.load_img('dog/dog.jpg',target_size=(200,200))
```

```python
type(img)
```

```python
'''
Convert to Numpy Array
'''
img=image.img_to_array(img)
```

```python
type(img)
```

```python
img.shape
```

```python
'''
Allocating Batch size=1
'''
input_batch=img.reshape(1,200,200,3)
```

```python
'''
Creating Generation Function
'''
datagen=ImageDataGenerator(
    rotation_range=30,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    width_shift_range=0.2,
    height_shift_range=0.2,
    fill_mode='reflect'
)
```

```python
'''
Generating image, saved into aug_dog
'''
i=0
for output in datagen.flow(input_batch,batch_size=1,save_to_dir='aug_dog'):
    i=i+1
    if i==10:
        break
```
